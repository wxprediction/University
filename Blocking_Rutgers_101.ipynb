# Percentile

```python
print()
print('Daily Values')
print("  0th   Percentile = " "%5.1f" % np.nanpercentile(data['Value'],0))
print("  5th   Percentile = " "%5.1f" % np.nanpercentile(data['Value'],5))
print("  20th  Percentile = " "%5.1f" % np.nanpercentile(data['Value'],20))
print("  50th  Percentile = " "%5.1f" % np.nanpercentile(data['Value'],50))
print("  80th  Percentile = " "%5.1f" % np.nanpercentile(data['Value'],80))
print("  95th  Percentile = " "%5.1f" % np.nanpercentile(data['Value'],95))
print("  100th Percentile = " "%5.1f" % np.nanpercentile(data['Value'],100))
```

# Plot the Mean - Annual and Seasonal

```python
fig, ax = plt.subplots(1,0,figsize=(12,6), dpi=90) 
plt.plot(data.Value)
plt.title('AO Daily Data - All Months - All Years')


fig, ax = plt.subplots(1,0,figsize=(12,6), dpi=90) 
plt.plot(data.Value[(data.Month==11) & (data.Day==30)])
plt.title('AO Daily Data - November - All Years')

 
```
